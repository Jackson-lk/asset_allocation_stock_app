```python
import pandas as pd
```

```python
df = pd.read_csv('../../../Dataset/stats_final_mean.csv')
```

```python
df
```

```python
Growth_of_Company = 3
Value_of_Company = 3
PE_ratio = 2
Return = 4
Stock_Volatility = 1
Stock_Volatility_to_Market = 2
```

```python
df_result = df.copy()
df_result['Result'] =""
df_result
```

```python
for i in range(0, len(df.index)):
    df_result.loc[i, 'Revenue_Growth'] = (df.loc[i, 'Revenue_Growth']*Growth_of_Company)
    df_result.loc[i, 'Enterprise_Value'] = (df.loc[i, 'Enterprise_Value']*Value_of_Company)
    df_result.loc[i, 'Forward_P/E'] = (df.loc[i, 'Forward_P/E']*PE_ratio)
    df_result.loc[i, 'Risk_Free_Return'] = (df.loc[i, 'Risk_Free_Return']*Return)
    df_result.loc[i, 'Stock_Volatility'] = (df.loc[i, 'Stock_Volatility']*Stock_Volatility*-1)
    df_result.loc[i, 'Beta'] = (df.loc[i, 'Beta']*Stock_Volatility_to_Market*-1)
    df_result.loc[i, 'Result'] = (df_result.loc[i, 'Revenue_Growth'] + df_result.loc[i, 'Enterprise_Value'] + df_result.loc[i, 'Forward_P/E'] + df_result.loc[i, 'Risk_Free_Return'] + df_result.loc[i, 'Stock_Volatility'] +  df_result.loc[i, 'Beta'])
df_result
```

```python
df_ticker = df_result.sort_values('Result', ascending = False).head(5)
df_ticker
```
